```python
data.dropna(inplace=True)
```

```python
data.value_counts('city')
```

```python
data['city'].value_counts().plot(kind='bar',color=['#CCB974','#8172B3','#64B5CD'])
plt.grid(ls=':',c='k')
plt.xticks(rotation=360)
plt.show()
```

```python
sns.histplot(x="iq", hue="placement", data=data, multiple="stack")
```

```python
figsize = (12, 1.2 * len(data['city'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(data, x='cgpa', y='city', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
plt.show()
```

```python
figsize = (12, 1.2 * len(data['city'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(data, x='iq', y='city', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
plt.show()
```

```python
sns.scatterplot(x = 'cgpa', y = 'iq', hue = 'city', data = data)
plt.show()
```

```python
data['cgpa'].plot(kind='hist', bins=20, title='cgpa')
plt.gca().spines[['top', 'right',]].set_visible(False)
plt.show()
```

```python
data.plot(kind='scatter', x='cgpa', y='iq', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```
